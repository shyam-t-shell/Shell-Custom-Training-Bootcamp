```python
temperature_data = [25.3, 26.1, 24.8, 23.5, 27.2]
pressure_data = [101.2, 100.8, 101.5, 100.2, 101.0]
humidity_data = [55.2, 54.8, 56.5, 53.7, 55.9]

temperature_arr=np.array(temperature_data)
pressure_arr=np.array(pressure_data)
humidity_arr=np.array(humidity_data)
```

```python
print("Temperature Mean:", temperature_arr.mean())
print("Pressure Mean:", pressure_arr.mean())
print("Humidity Mean:", humidity_arr.mean())

print("Temperature Standard Deviation:", temperature_arr.std())
print("Pressure Standard Deviation:", pressure_arr.std())
print("Humidity Standard Deviation:", humidity_arr.std())
```

```python
temperature_mean=np.mean(temperature_arr)
pressure_mean=np.mean(pressure_arr)
humidity_mean=np.mean(humidity_arr)

temperature_std=np.std(temperature_arr)
pressure_std=np.std(pressure_arr)
humidity_std=np.std(humidity_arr)

temperature_max=np.max(temperature_arr)
pressure_max=np.max(pressure_arr)
humidity_max=np.max(humidity_arr)

temperature_min=np.min(temperature_arr)
pressure_min=np.min(pressure_arr)
humidity_min=np.min(humidity_arr)



print("Mean Temperature:", temperature_mean)
print("Std Temperature:", temperature_std)
print("Max Temperature:", temperature_max)
print("Min Temperature:", temperature_min)

print("Mean Pressure:", pressure_mean)
print("Std Pressure:", pressure_std)
print("Max Pressure:", pressure_max)
print("Min Pressure:", pressure_min)

print("Mean Humidity:", humidity_mean)
print("Std Humidity:", humidity_std)
print("Max Humidity:", humidity_max)
print("Min Humidity:", humidity_min)
```

```python
#Create a list of months: ['January', 'February', 'March', 'April', 'May'].
#Create a list of monthly sales data for a hypothetical product: [2500, 3200, 2800, 4100, 3700]

months = np.array(['January', 'February', 'March', 'April', 'May'])
sales = np.array([2500, 3200, 2800, 4100, 3700])

plt.figure(figsize=(5,5))
plt.bar(months,sales, color='orange', width=0.5)
plt.xlabel("Months", fontsize=10)
plt.ylabel("Sales Amount", fontsize=10)
plt.title("Sales in various months", fontsize=15)
plt.savefig('Monthly Sales.png')
```
