### Recursive Functions

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

factorial(3)
```

```python
def fibo(n):
    if(n==1):
        return 0
    elif(n==2):
        return 1
    else:
        return fibo(n-1)+fibo(n-2) 
fibo(8)
```

### Decorator Function

```python
def og_div(a,b):
    c=a/b
    return c
```

```python
def smart_div(func):
    def inner(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return inner
```

```python
result = smart_div(og_div)
```

```python
result(10,5)
```

```python
def wish(name):
    print("Hello", name, "Good Morning")
    
def wish_specific(func):
    def inner(n):
        if n=="Robert":
            print("Hello", n, "Bad Morning")
        else:
            func(n)
    return inner

result = wish_specific(wish)
    
result("John")
result("Steve")
result("Robert")
```

```python
def wish_specific(func):
    def inner(n):
        if n=="Robert":
            print("Hello", n, "Bad Morning")
        else:
            func(n)
    return inner

@wish_specific
def wish(name):
    print("Hello", name, "Good Morning")
    
result("John")
result("Steve")
result("Robert")
```

### Generator Function
A Generator Functionis yield keyword to yield a sequence of values one at a time, instead of returning all at once.

```python
def count_up_to(n):
    i=1
    while i<=n:
        yield i
        i=i+1
```

```python
count_up_to(10)
```

```python
for i in count_up_to(10):
    print(i)
```

```python
# to generate fibonacci series using generator function
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib_series = []
for _ in range(20):
    fib_series.append(next(fibonacci_generator()))

print(fib_series)
```
