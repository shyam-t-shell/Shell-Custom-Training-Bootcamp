```python
import pyspark

import findspark
findspark.init('/usr/local/spark')

from pyspark import SparkContext

conf = pyspark.SparkConf().setMaster("local").setAppName("first")
sc = SparkContext(conf=conf)
```

```python
rdd=sc.parallelize([1,2,3])
```

```python
rdd.collect()
```

```python
sc
```

```python
rdd2=sc.parallelize(["Python","SQL","PySpark"])
display(rdd2)
display(rdd2.collect())
display(type(rdd2))
```

```python
rdd3=sc.parallelize([1,2,3,4,5,6,7,8])
rdd3.collect()
```

```python
# Make the elements twice inside the list twice of the RDD
# [2,4,6,8,10,12,14,16]
rdd4 = rdd3.map(lambda x:x*2)
rdd4.collect()
```

```python
# Save all the even numbers in new rdd
rdd5 = rdd3.filter(lambda x: x%2==0)
rdd5.collect()
```

```python
from datetime import datetime, date
import pandas as pd
from pyspark.sql import Row

df = spark.createDataFrame([
    Row(a=1, b=2., c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
    Row(a=2, b=3., c='string2', d=date(2000, 2, 1), e=datetime(2000, 1, 2, 12, 0)),
    Row(a=4, b=5., c='string3', d=date(2000, 3, 1), e=datetime(2000, 1, 3, 12, 0))
])

df
```

```python
df.show()
```

```python
schema = 'a long, b double, c string, d date, e timestamp'
data = [
    (1, 2., 'string1', date(2000, 1, 1), datetime(2000, 1, 1, 12, 0)),
    (2, 3., 'string2', date(2000, 2, 1), datetime(2000, 1, 2, 12, 0)),
    (3, 4., 'string3', date(2000, 3, 1), datetime(2000, 1, 3, 12, 0))
]

df = spark.createDataFrame(data,schema)
df
```

```python
df.show()
```

```python
pandas_df = pd.DataFrame({
    'a': [1, 2, 3],
    'b': [2., 3., 4.],
    'c': ['string1', 'string2', 'string3'],
    'd': [date(2000, 1, 1), date(2000, 2, 1), date(2000, 3, 1)],
    'e': [datetime(2000, 1, 1, 12, 0), datetime(2000, 1, 2, 12, 0), datetime(2000, 1, 3, 12, 0)]
})

df = spark.createDataFrame(pandas_df)
df
```

```python
df = spark.read.option("header", True).option("inferschema", True).csv('employees.csv')
df.printSchema()
```

```python
df.select("First Name","Gender").show(4)
```

```python
# df.select("First Name".alias("Name"))
# Error Received: (AttributeError: 'str' object has no attribute 'alias')

# Imp to know difference between column type and string type
from pyspark.sql.functions import col
df.select(col("First Name").alias("Name")).show(5)
```

```python
df.select("First Name", col("Gender"), df["Team"]).show(5)
```

```python
df.withColumnRenamed('First Name', 'Name').show(5)
```

```python
from pyspark.sql.functions import *
df.select(concat('Start Date',lit(' '),'Last Login Time')).show(5)
```

```python
from pyspark.sql.functions import *
df.select(
    "*",
    concat('Start Date',lit(' '),'Last Login Time').alias("DateTime")
).show(5)
```

```python
from pyspark.sql.functions import *
df1 = df.select("*",concat('Start Date',lit(' '),'Last Login Time').alias("DateTime"))
df1 = df1.drop("Start Date", "Last Login Time")
df1.show(5)
```

```python
df1.write.csv("Spark_Final_Dataframe.csv")
```
